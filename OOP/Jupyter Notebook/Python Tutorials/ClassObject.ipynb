## **Lớp và đối tượng trong Python**

### **Lớp và đối tượng trong Python**

#### **Lớp (class) trong Python**

**Một lớp là một bản thiết kế cho đối tượng.** Chúng ta có thể coi một lớp như một bản phác thảo của một ngôi nhà với các chi tiết. Nó chứa tất cả các chi tiết về tên, màu sắc, kích thước,.. Dựa trên những mô tả này, chúng ta có thể xây dựng các ngôi nhà.

Ở đây, **một ngôi nhà là một đối tượng**. Tức là bạn có thể sử dụng bản thiết kế (hay lớp này) để định nghĩa bao nhiêu đối tượng tùy thích.

#### **Đối tượng (object) trong Python**

Một đối tượng là một thể hiện của một lớp. Khi lớp được định nghĩa, chỉ có mô tả cho đối tượng được định nghĩa. Do đó, **không có bộ nhớ nào được cấp phát.** Khi đó, nếu chúng ta muốn sử dụng lớp đó, chúng ta cần cấp phát bộ nhớ nhằm sử dụng nó. Hay nói cách khác là gọi đến hàm tạo của nó.

Về cơ bản, các đối tượng có các tính chất sau đây:

- **Trạng thái hay thuộc tính:** Một đối tượng có các thuộc tính đại diện cho các trạng thái của nó, và nó phản ánh các đặc tính của đối tượng. Ví dụ: màu xanh, màu đỏ, cân nặng,...

- **Hành vi hay phương thức:** Một đối tượng sẽ có các phương thức để thể hiện các hành vi hoặc hành động nào đó trong chương trình. Ví dụ: đọc, ghi, nói,....

**Ví dụ:** Nếu chúng ta thiết kế một lớp trên các trạng thái và hành vi của một Người, thì các Trạng thái có thể được biểu diễn dưới dạng các biến thể hiện đặc điểm của người đó và các hành vi dưới dạng các phương thức của lớp.

Ở trên là một ví dụ thực tế với lớp và đối tượng. Ở đây chúng ta có 1 lớp ***Person***, trong đó:

**- Có các thuộc tính:** Tên, Giới tính, Nghề nghiệp

**- Và các hành vi:** Đi làm, Đi học

Sử dụng các thuộc tính và hành vi này, chúng ta có thể tạo thành rất nhiều các đối tượng khác nhau:

- Đối tượng 1 có các thuộc tính: 
    - Tên là Hùng, 
    - Giới tính: Nam, 
    - Nghề nghiệp: Lập trình viên; 
    - và có hành vi là: Đi làm lập trình viên tại TEK4.VN và mỗi ngày Đi học 2 tiếng





- Đối tượng 2 có các thuộc tính: 
    - Tên là Giang, 
    - Giới tính: Nam, 
    - Nghề nghiệp: Bác sĩ; 
    - và có các hành vi là: Đi làm tại bệnh viện Bạch Mai, và mỗi ngày Đi học 5 tiếng.

Chúng ta có thể thấy lớp được định nghĩa ở trên là giống nhau nhưng các đối tượng được tạo ra từ cùng một lớp có thể khác nhau. Chúng có thể có các đặc tính và hành vi khác nhau.

#### **Định nghĩa một lớp trong Python**

```python
class TenLop:
# docstrings
    <Câu lệnh 1>
    <Câu lệnh 2>
    .
    .
    <Câu lệnh N>
```

**Trong đó:**

- TenLop là tên được đặt cho lớp được định nghĩa, chúng ta sẽ tương tác với các đối tượng sinh ra từ lớp bằng cách sử dụng tên này.

- Chuỗi đầu tiên bên trong lớp được gọi là docstring và có nhiệm vụ mô tả ngắn gọn về lớp, mặc dù là không bắt buộc.

- Các câu lệnh để định nghĩa các thuộc tính và hành vi (phương thức) của đối tượng trong lớp.


Về mặt biểu diễn các khối lệnh, chúng ta cũng áp dụng các quy tắc thụt lề tương tự như hàm với lưu ý ở đây khối định nghĩa lớp là khối lệnh lớn nhất, sau đó sẽ đến các khối lệnh định nghĩa hàm (phương thức) trong lớp, các khối điều kiện, vòng lặp...Mỗi khối lệnh nhỏ nằm trong khối lệnh lớn sẽ phải thụt dòng một khoảng bằng nhau!

Ví dụ, đoạn chương trình sau định nghĩa một lớp Person:

```python
class SinhVien:
     "Đây là docstring của lớp SinhVien"
     pass
```

Thông thường để code nhìn trong sáng, chúng ta nên đặt tên lớp theo quy tắc đó là **tên lớp nên bắt đầu bằng chữ viết Hoa**. Nếu tên lớp chứa nhiều từ, chúng ta nên sử dụng cú pháp dạng lạc đà  (CamelCase), tức là viết hoa các chữ cái đầu của mỗi từ. Chẳng hạn **OfficeManager, FulltimeEmployee, SinhVien,...**

Ở ví dụ trên, chúng ta chưa biết làm gì với lớp này nên chúng ta sử dụng từ khóa ***pass*** để bỏ qua các đoạn mã bắt buộc để tránh chương trình báo lỗi. Điều này giúp giữ chỗ để thêm các đoạn mã khác sau này.

#### **Tạo một đối tượng của lớp trong Python**

Để tạo một đối tượng của lớp, chúng ta có thể đơn giản khai báo một biến và gán cho nó tên lớp theo sau bởi dấu ngoặc tròn như sau:

```python
a = SinhVien()
b = SinhVien()
```

Ở đây, chúng ta đã tạo các đối tượng a và b là các thể hiện của lớp **SinhVien** trong chương trình bằng cách khai báo và gán nó với hàm tạo lớp là **SinhVien()**. 

Chắc hẳn trước đây bạn cũng nhớ đến các hàm tạo danh sách **list()**, hàm tạo từ điển **dict(),**...**Tất cả những hàm này là các hàm tạo của lớp.**

```python
class SinhVien:
     "Đây là docstring của lớp SinhVien"

     # State
     # Behavior
     pass

a = SinhVien()
b = SinhVien()
print(a)
print(b)
```

Chúng ta sẽ thu được địa chỉ của chúng trong bộ nhớ

#### **Phương thức của đối tượng**

Như đã đề cập ở trên, một đối tượng sẽ có các hành vi. **Các hành vi này sẽ được mô hình hóa trong Python thành các hàm**.

Khi các ***hàm thuộc vào một lớp***, nó sẽ được gọi là ***các phương thức***. ***Những điều chúng ta đã học với hàm cũng được áp dụng y hệt như với phương thức***.

```python
class SinhVien:
    # phuong thuc
    def in_thong_tin(self):
        print('Sinh viên')
```

Trong đoạn mã trên, chúng ta đã định nghĩa một lớp **SinhVien** với phương thức ***in_thong_tin()***. Phương thức này nhìn giống với các hàm thông thường trừ việc nó có một tham số **self**, chúng ta sẽ tìm hiểu về tham số self này phần tiếp sau dưới đây.

Để sử dụng một phương thức, chúng ta thực hiện như sau:

```python
object_name.method_name()
```

Ở đây, chúng ta xét một ví dụ với lớp **SinhVien** ở trên. Chúng ta sẽ tạo ra một đối tượng a và gọi phương thức **in_thong_tin** trên đối tượng **a** này:

```python
class SinhVien:
     def in_thong_tin(self):
          print("Sinh viên")

a = SinhVien()
a.in_thong_tin()
```

Chúng ta thấy rằng ở ví dụ trên có một điều khá lạ so với trước đây chúng ta được học. Đó là, hàm **in_thong_tin** được khai báo có một tham số nhưng khi sử dụng chúng ta lại **không đưa vào bất cứ tham số nào, nhưng hàm vẫn chạy đúng** !!!

Ví dụ nếu ta sử dụng như sau:

```python
def in_thong_tin(self):
     print("Sinh viên")

in_thong_tin()
```

Chương trình sẽ báo lỗi, Tuy nhiên, với phương thức trong một lớp thì điều này không vấn đề gì.

Lí do đó là khi chúng ta gọi một phương thức trên một đối tượng, Python đã tự động truyền đối tượng đó vào phương thức như là tham số đầu tiên của phương thức này. Tham số ***self*** thực chất là chỉ chính đối tượng của lớp gọi đến phương thức đó. Trong ví dụ trên, khi đối tượng ***a gọi đến phương thức in_thong_tin()***:

```python
a.in_thong_tin()
```

```python
# Hay thực chất nó là:
a.in_thong_tin(a)
```

Nếu bạn không tin self chính là a. Vậy thì hãy thử đoạn chương trình sau đây:

```python
class SinhVien:
     def in_thong_tin(self):
          print(self)
          print("Sinh viên")

a=SinhVien()
a.in_thong_tin()
print(a)
```

Ta có thể thấy, địa chỉ của ***self và a*** là như nhau, hay nói cách khác hai đối tượng này tham chiếu đến cùng một đối tượng và tương đương nhau.

***Nguyên tắc ở đây là: tất cả các phương thức của một lớp đều phải có ít nhất một tham số (self) là đối tượng gọi đến phương thức đó.***

----

**Tip&Trick**

**Lưu ý: self không phải là từ khóa, chúng ta có thể đặt tham số trong hàm là bất cứ thứ gì. Khi đó tham số đầu tiên sẽ được hiểu có tác dụng như self tức là tham chiếu đến đối tượng gọi phương thức. Tuy nhiên, chúng ta không nên sử dụng như vậy bởi sẽ gây khó hiểu và không tường minh khi đọc và sử dụng.**
