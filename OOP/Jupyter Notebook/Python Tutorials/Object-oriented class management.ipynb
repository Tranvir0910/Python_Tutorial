#### **Bc 1. nh ngha cc thuc tnh ca i tng**

iu u tin chng ta cn thc hin l nh ngha cc lp c trong chng trnh. V trong chng trnh c 3 ch th l kha hc, ging vin v sinh vin. Do , chng ta s nh ngha 3 lp bao gm **Course**, **Lecturer** v **Student** nh sau:

**Lp Course:**

```python
class Course:
    def __init__(self, name,year,id_number,lecturer):
        self.name = name
        self.year = year
        self.id_number = id_number
        self.lecturer = lecturer
        self.student_list = []
```

  y, ch  rng, chng ta cha c danh sch sinh vin  tin hnh khi to, do ,  y chng ta khi to n l mt danh sch rng ([])

**Lp Student v Lecturer:**

Chng ta  thy  cc bi trc rng, lp Student v Lecturer l gn ging nhau v mt nh ngha. Do ,  rt gn on m nh ngha 2 lp ny bng khi nim k tha trong Python. Chng ta s nh ngha li thm mt lp l Person v cho lp Student, Lecturer k tha t lp Person ny nh sau:

```python
class Person:
    def __init__(self, first_name, last_name, age, email):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email

class Lecturer(Person):
    def __init__(self,f_name,l_name,age,email,bank_account):
        super().__init__(f_name, l_name, age, email)
        self.bank_account = bank_account
       
class Student(Person):
    def __init__(self,f_name,l_name,age,email,student_id, grade=-1):
        super().__init__(f_name, l_name, age, email)
        self.student_id = student_id
        self.grade = grade
```

#### **Bc 2. nh ngha cc phng thc ca i tng**

Trong chng trnh yu cu cho php sinh vin ng k v in ra thng tin ca kha hc cng vi thng tin v ging vin v danh sch sinh vin. Khi  trong lp Course, chng ta cn nh ngha thm cc phng thc: **print_student()**  in ra danh sch sinh vin, **print_lecturer()**  in ra thng tin ging vin:

```python
# Thuc lp Course

# Danh sch sinh vin
def enroll(self,student):
    self.student_list.append(student)

# Thng tin hc sinh
def print_student(self):
    print("----------Students in course " + self.name + "----------")
    for i in self.student_list:
        txt = "{:<16} {:<16} {:<6d} {:<16}"
        txt2 = "{:<16} {:<16} {:<6} {:<16}"
        print(txt2.format("First name", "Last name", "Age", "Email"))
        print(txt.format(i.first_name, i.last_name, i.age, i.email))

# Thng tin ging vin
def print_lecturer(self):
    print("\n\n\n----------Lecturer in course " + self.name + "----------")
    txt = "{:<16} {:<16d}"
    txt2 = "{:<16} {:<16}"
    print(txt2.format("Name", "Bank Account"))
    print(txt.format(self.lecturer.first_name, self.lecturer.bank_account))
```

i vi 2 i tng Ging vin v Sinh vin, chng ta mun in ra thng tin ca cc i tng ny. Chng ta c mt s thng tin chung. V vy, chng ta s to mt ***phng thc in cc thng tin chung ny trong lp Person  cho php cc lp Lecturer v Student k tha n***:

```python
# Thuc lp Person
def print_info(self):
    print(self.first_name +" " +self.last_name +" is " + str(self.age) +" years old.")
```

```python
def print_info(self):
    super().print_info()
    print("Student ID: " + self.student_id)
    print("Grade: " + self.grade)
```

Cui cng, ton b on m ca chng trnh c th hin nh di y:

```python
class Course:
    def __init__(self, name, year, id_number, lecturer):
        self.name = name
        self.year = year
        self.id_number = id_number
        self.lecturer = lecturer
        self.student_list = []

    def enroll(self,student):
        self.student_list.append(student)

    def print_student(self):
        print("----------Students in course " + self.name + "----------")
        for i in self.student_list:
            txt = "{:<16} {:<16} {:<6d} {:<16}"
            txt2 = "{:<16} {:<16} {:<6} {:<16}"
            print(txt2.format("First name", "Last name", "Age", "Email"))
            print(txt.format(i.first_name, i.last_name, i.age, i.email))
           
    def print_lecturer(self):
        print("\n\n\n----------Lecturer in course " + self.name + "----------")
        txt = "{:<16} {:<16d}"
        txt2 = "{:<16} {:<16}"
        print(txt2.format("Name", "Bank Account"))
        print(txt.format(self.lecturer.first_name, self.lecturer.bank_account))

class Person:
    def __init__(self, first_name,last_name,age,email):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email
    def print_info(self):
        print(self.first_name +" " +self.last_name +" is " + str(self.age) +" years old.")
        print("Email:" + self.email)

class Lecturer(Person):
    def __init__(self,f_name,l_name,age,email,bank_account):
        super().__init__(f_name,l_name,age,email)
        self.bank_account = bank_account
    
    def print_info(self):
        super().print_info()
        print("Bank Accout:" + str(self.bank_account))
       
class Student(Person):
    def __init__(self,f_name,l_name,age,email,student_id, grade=-1):
        super().__init__(f_name,l_name,age,email)
        self.student_id = student_id
        self.grade=grade
    
    def print_info(self):
        super().print_info()
        print("Student ID: " + self.student_id)
        print("Grade: " + str(self.grade))


#Khi to cc i tng
lecturer_TEK4VN = Lecturer("Hng", "ng", 35, "hungnguyen@tek4.vn", 1234567)
python_course = Course("Lp trnh Python c bn", 2019, 1234, lecturer_TEK4VN)
student_hai = Student("Hi","Nguyn", 18, "hainguyen@tek4.vn", "108001024")
student_ba=Student("Ba", "Nguyn", 21, "banguyen@tek4.vn", "108001025")
python_course.enroll(student_hai)
python_course.enroll(student_ba)

#In ra sinh vin c trong kha hc
python_course.print_student()
python_course.print_lecturer()
student_hai.print_info()
lecturer_TEK4VN.print_info()
```
