Các thuộc tính này là khác nhau với mỗi đối tượng, do đó chúng ta sẽ không thể sử dụng chung thuộc tính nào. Các thuộc tính đều cần khởi tạo để nhận giá trị tương ứng khi tạo một đối tượng khóa học từ lớp Course. 

Do vậy, chúng ta sẽ đưa tất cả các thuộc tính vào khởi tạo trong hàm tạo của lớp là __ init__():

```python
class Course:
    def __init__(self, name, year, id_number, lecturer, student):
        self.name = name
        self.year = year
        self.id_number = id_number
        self.lecturer = lecturer
        self.student_list = student
```

```python
# Phương thức
def print_info(self):
    print(self.name + " " + self.year + " " + self.id_number + " " + self.lecturer)
    print(self.student_list)
```

Sẽ thế nào nếu như có nhiều giảng viên hay sinh viên trùng tên?

Trong bài này, chúng ta sẽ giải quyết vấn đề cho trường hợp này bằng cách sử dụng các đối tượng giảng viên (Lecturer) và sinh viên (Student). Khi đó mỗi giảng viên và sinh viên sẽ được đại diện bằng 1 đối tượng các các thông tin khác nhau. Ít nhất ở trường bank_account (với giảng viên) hoặc student_id (với sinh viên). Khi đó, chúng ta sẽ không lo trường hợp dữ liệu bị trùng lặp nữa.

Với lớp Lecturer, chúng ta sẽ định nghĩa các thuộc tính của mỗi đối tượng thuộc lớp này bao gồm: tên (first_name), họ (last_name), tuổi (age), địa chỉ email (email) và tài khoản ngân hàng (bank_account):

```python
class Lecturer:
    def __init__(self, first_name,last_name,age,email, bank_account):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email
        self.bank_account = bank_account
```

Tương tự như vậy, chúng ta sẽ định nghĩa lớp Student bao gồm các thuộc tính: tên (first_name), họ (last_name), tuổi (age), địa chỉ email (email), mã số sinh viên (student_id) và điểm mặc định có giá trị là -1 (grade=-1) như sau:

```python
class Student:
    def __init__(self, first_name,last_name,age,email, student_id):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email
        self.student_id = student_id
        self.grade = -1
```

Để in thông tin cho các đối tượng trong hai lớp, chúng ta khai báo thêm một phương thức print_info() cho hai lớp này:

```python
def print_info(self):
    print(self.first_name +" " +self.last_name +" is " + str(self.age) +" years old.")
```

Cuối cùng, chúng ta sẽ thử khai báo một vài đối tượng của hai lớp này để thấy hoạt động của nó:

```python
lecturer_tung = Lecturer("Tùng","Phạm",35,"tungpham@tek4.vn","1234567")
student_hai = Student("Hải","Nguyễn",21,"hainguyen@tek4.vn","108001024")

lecturer_tung.print_info()
student_hai.print_info()
```

### **Hoàn thiện chương trình quản lý khóa học**

#### **Bước 1. Định nghĩa các thuộc tính của đối tượng**

Điều đầu tiên chúng ta cần thực hiện là định nghĩa các lớp có trong chương trình. Vì trong chương trình có 3 chủ thể là khóa học, giảng viên và sinh viên. Do đó, chúng ta sẽ định nghĩa 3 lớp bao gồm **Course**, **Lecturer** và **Student** như sau:

**Lớp Course:**

```python
class Course:
    def __init__(self, name,year,id_number,lecturer):
        self.name = name
        self.year = year
        self.id_number = id_number
        self.lecturer = lecturer
        self.student_list = []
```

 Ở đây, chú ý rằng, chúng ta chưa có danh sách sinh viên để tiến hành khởi tạo, do đó, ở đây chúng ta khởi tạo nó là một danh sách rỗng ([])

**Lớp Student và Lecturer:**

Chúng ta đã thấy ở các bài trước rằng, lớp Student và Lecturer là gần giống nhau về mặt định nghĩa. Do đó, để rút gọn đoạn mã định nghĩa 2 lớp này bằng khái niệm kế thừa trong Python. Chúng ta sẽ định nghĩa lại thêm một lớp là Person và cho lớp Student, Lecturer kế thừa từ lớp Person này như sau:

```python
class Person:
    def __init__(self, first_name, last_name, age, email):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email

class Lecturer(Person):
    def __init__(self,f_name,l_name,age,email,bank_account):
        super().__init__(f_name, l_name, age, email)
        self.bank_account = bank_account
       
class Student(Person):
    def __init__(self,f_name,l_name,age,email,student_id, grade=-1):
        super().__init__(f_name, l_name, age, email)
        self.student_id = student_id
        self.grade = grade
```

#### **Bước 2. Định nghĩa các phương thức của đối tượng**

Trong chương trình yêu cầu cho phép sinh viên đăng ký và in ra thông tin của khóa học cùng với thông tin về giảng viên và danh sách sinh viên. Khi đó trong lớp Course, chúng ta cần định nghĩa thêm các phương thức: **print_student()** để in ra danh sách sinh viên, **print_lecturer()** để in ra thông tin giảng viên:

```python
# Thuộc lớp Course

# Danh sách sinh viên
def enroll(self,student):
    self.student_list.append(student)

# Thông tin học sinh
def print_student(self):
    print("----------Students in course " + self.name + "----------")
    for i in self.student_list:
        txt = "{:<16} {:<16} {:<6d} {:<16}"
        txt2 = "{:<16} {:<16} {:<6} {:<16}"
        print(txt2.format("First name", "Last name", "Age", "Email"))
        print(txt.format(i.first_name, i.last_name, i.age, i.email))

# Thông tin giảng viên
def print_lecturer(self):
    print("\n\n\n----------Lecturer in course " + self.name + "----------")
    txt = "{:<16} {:<16d}"
    txt2 = "{:<16} {:<16}"
    print(txt2.format("Name", "Bank Account"))
    print(txt.format(self.lecturer.first_name, self.lecturer.bank_account))
```

Đối với 2 đối tượng Giảng viên và Sinh viên, chúng ta muốn in ra thông tin của các đối tượng này. Chúng ta có một số thông tin chung. Vì vậy, chúng ta sẽ tạo một ***phương thức in các thông tin chung này trong lớp Person để cho phép các lớp Lecturer và Student kế thừa nó***:

```python
# Thuộc lớp Person
def print_info(self):
    print(self.first_name +" " +self.last_name +" is " + str(self.age) +" years old.")
```

Cuối cùng, lớp Student kế thừa lớp Person để thực hiện in ra thông tin:

```python
def print_info(self):
    super().print_info()
    print("Student ID: " + self.student_id)
    print("Grade: " + self.grade)
```

Và tương tự với lớp Lecturer:

```python
def print_info(self):
    super().print_info()
    print("Bank Accout:" + self.bank_account)
```

Ở đây, chúng ta đã thực hiện việc kế thừa hàm print_info() từ lớp Person, và chúng ta không muốn phức tạp dùng nhiều tên hàm khác nhau, do đó, chúng ta sử dụng khái niệm đa hình (trong trường hợp này là nạp chồng) để định nghĩa hai phương thức print_info() khác trong hai lớp Student và Lecturer.

```python
class Course:
    def __init__(self, name, year, id_number, lecturer):
        self.name = name
        self.year = year
        self.id_number = id_number
        self.lecturer = lecturer
        self.student_list = []

    def enroll(self,student):
        self.student_list.append(student)

    def print_student(self):
        print("----------Students in course " + self.name + "----------")
        for i in self.student_list:
            txt = "{:<16} {:<16} {:<6d} {:<16}"
            txt2 = "{:<16} {:<16} {:<6} {:<16}"
            print(txt2.format("First name", "Last name", "Age", "Email"))
            print(txt.format(i.first_name, i.last_name, i.age, i.email))
           
    def print_lecturer(self):
        print("\n\n\n----------Lecturer in course " + self.name + "----------")
        txt = "{:<16} {:<16d}"
        txt2 = "{:<16} {:<16}"
        print(txt2.format("Name", "Bank Account"))
        print(txt.format(self.lecturer.first_name, self.lecturer.bank_account))

class Person:
    def __init__(self, first_name,last_name,age,email):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email
    def print_info(self):
        print(self.first_name +" " +self.last_name +" is " + str(self.age) +" years old.")
        print("Email:" + self.email)

class Lecturer(Person):
    def __init__(self,f_name,l_name,age,email,bank_account):
        super().__init__(f_name,l_name,age,email)
        self.bank_account = bank_account
    
    def print_info(self):
        super().print_info()
        print("Bank Accout:" + str(self.bank_account))
       
class Student(Person):
    def __init__(self,f_name,l_name,age,email,student_id, grade=-1):
        super().__init__(f_name,l_name,age,email)
        self.student_id = student_id
        self.grade=grade
    
    def print_info(self):
        super().print_info()
        print("Student ID: " + self.student_id)
        print("Grade: " + str(self.grade))


#Khởi tạo các đối tượng
lecturer_TEK4VN = Lecturer("Hưng", "Đặng", 35, "hungnguyen@tek4.vn", 1234567)
python_course = Course("Lập trình Python cơ bản", 2019, 1234, lecturer_TEK4VN)
student_hai = Student("Hải","Nguyễn", 18, "hainguyen@tek4.vn", "108001024")
student_ba=Student("Ba", "Nguyễn", 21, "banguyen@tek4.vn", "108001025")
python_course.enroll(student_hai)
python_course.enroll(student_ba)

#In ra sinh viên có trong khóa học
python_course.print_student()
python_course.print_lecturer()
student_hai.print_info()
lecturer_TEK4VN.print_info()
```
