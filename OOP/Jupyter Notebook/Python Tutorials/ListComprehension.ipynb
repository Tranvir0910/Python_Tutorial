## **Kỹ thuật Comprehension trong Python**

### **1. List Comprehension trong Python**

Kỹ thuật, **List comprehension** trong Python cho phép thay vì phải tạo một danh sách rỗng và thêm từng phần tử vào cuối, chúng ta chỉ cần **định nghĩa cho tập hợp đó và nội dung của nó cùng một lúc** bằng cách làm theo định dạng sau:

```python
output_List = [expression(variable) for variable in input_List [if variable condition1][if variable condition2]] 
```

#### **Trong đó:**

**1. expression(variable) (không bắt buộc)**: là một câu lệnh gọi đến một phương thức hoặc bất kỳ biểu thức hợp lệ nào khác trả về một giá trị.

**2. variable (bắt buộc)**:là đối tượng hoặc giá trị trong tập hợp kiểu danh sách hoặc đối tượng có thể duyệt trong danh sách đầu vào **input_List**

**3. input_List (bắt buộc)**: là tập hợp danh sách đầu vào để duyệt hoặc bất kỳ đối tượng nào khác có thể trả về từng phần tử của nó tại một thời điểm.

**4. condition1, condition2 ... (không bắt buộc)**: Là các điều kiện để lọc phần tử cần lấy ra cho nó thỏa mãn các điều kiện này (nếu có).

#### **Ví dụ 1:**


Giả sử rằng, chúng ta muồn tạo một danh sách đầu ra chỉ chứa các số chẵn lấy từ danh sách đầu vào. Chúng ta có thể thực hiện nó với một vòng lặp for như sau:

```python
input_list = range(20)
  
output_list = []
for var in input_list:
    if var % 2 == 0:
        output_list.append(var)
  
print("Output List using for loop:", output_list)
```


Chúng ta thấy rằng nó khá phức tạp và khó hiểu phải không ạ?

Đơn giản hơn, chúng ta chỉ cần duy nhất một câu lệnh gán với kỹ thuật **list compreshension:**


```python
input_list = range(20)
list_using_comp = [var for var in input_list if var % 2 == 0]
print("Output List using list comprehensions:", list_using_comp)
```

Cả 2 đoạn mã trên cùng cho ra kết quả là một danh sách gồm các số chẵn từ 0 đến 18:

Tuy nhiên, có thể thấy rằng, cách sử dụng list compresshension hiệu quả và gọn gàng hơn rất nhiều!


**Ở đây:**

**1. expression** chính là **var** (tức là ta muốn lấy nguyên giá trị mà ta duyệt qua, không có thay đổi gì nó cả)

**2. variable** chính là **var** như một biến để ta duyệt qua **input_list** gồm các số từ 0 đến 19.

**3.** Điều kiện duyệt ở đây là số mà ta duyệt qua phải chia hết cho 2.

#### **Ví dụ 2:**

Giả sử ta muốn tạo một danh sách chứa bình phương của tất cả các số từ 1 đến 9. Khi đó, chúng ta xem xét cả 2 cách sử dụng **vòng lặp for** và **list comprehension** như sau:

```python
output_list = []
for var in range(1, 10):
    output_list.append(var ** 2)
      
print("Output List using for loop:", output_list)
list_using_comp = [var**2 for var in range(1, 10)]
print("Output List using comprehension:", list_using_comp)
```

Ta thấy cách sử dụng list compreshension có vẻ trong sáng và dễ thực hiện hơn rất nhiều.

#### **Ví dụ 3**

Tính số tiền phải trả sau sai khi đã cộng với thuế VAT 8% của đơn hàng siêu thị.

Đoạn mã sử dụng kỹ thuật List compreshension như sau:

```python
number = [1.09, 23.56, 57.84, 4.56, 6.78]
rate = 0.08

def get_price_with_tax(tax):
     return tax * (1 + rate)

final_prices = [get_price_with_tax(i) for i in number]
print(sum(final_prices))
```

#### **Ví dụ 4:**

Chuẩn hóa giá cả trong cơ sở dữ liệu:

```python
original_prices = [3.12, -2.32, 12.12, 13.14, -6.43]
prices = [i if i > 0 else 0 for i in original_prices]
print(prices)
```

Trong ví dụ trên, chúng ta có một tập hợp danh sách các giá cả, do nhập sai do đó có rất nhiều giá trị âm, chúng ta có thể thực hiện thay thế các giá trị âm bằng 0 và giữ nguyên các giá trị dương theo như đoạn mã trên.

### **2. Áp dụng Comprehension cho từ điển và Set**

Chúng ta cũng có thể sử dụng kỹ thuật Comprehension cho đối tượng kiểu **Set** và **Dictionary.**

**Set Comprehension** sẽ gần giống chính xác như **List Comprehension** trong Python.

**Sự khác biệt** là **Set Comprehension** sẽ đảm bảo **kết quả đầu ra không có các giá trị trùng lặp nào**.

Chúng ta có thể áp dụng kỹ thuật **Set Comprehension** bằng cách **sử dụng dấu ngoặc nhọn {}** thay vì **cặp dấu ngoặc vuông như list.**

#### **Ví dụ 5:**

```python
quote = "Python is easy"
unique_vowels = {i for i in quote if i in 'aeiou'}
print(unique_vowels)
```

Không giống như đối tượng kiểu List hay danh sách, đối tượng kiểu Set sẽ **không đảm bảo rằng các phần tử sẽ được lưu trữ theo bất kỳ thứ tự cụ thể** nào. Đây là lý do tại sao các phần tử đầu tiên của đối tượng kiểu Set là a, mặc dù giá trị đầu tiên trong chuỗi ký tự là i.

Miễn là tập hợp dữ liệu đó cho phép chúng ta duyệt qua nó là chúng ta có thể sử dụng **comprehension**. Các tập như vậy sau này chúng ta sẽ gọi là các tập **iterable**.

Việc áp dụng kỹ thuật Comprehension cho đối tượng kiểu từ điển hay **Dictionary** cũng giống tương tự, tuy nhiên sẽ có thêm yêu cầu là thực hiện **thêm giá trị khóa** khi thực hiện áp dụng kỹ thuật này.

Đoạn mã sau sẽ in ra một từ điển với khóa là giá trị nguyên từ 0 đến 9 và giá trị tương ứng với mỗi khóa là bình phương của khóa đó:
