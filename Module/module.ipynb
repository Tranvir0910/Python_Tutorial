## **Module trong Python**

Trong Python, cũng như đa số các ngôn ngữ lập trình khác, khi viết chương trình, chúng ta hiếm khi chỉ sử dụng một file mã nguồn duy nhất. Trên thực tế, chương trình là tổng hợp của rất nhiều file và thư mục phức tạp. Python quản lý các file và thư mục này dưới dạng các module và package.

Về cơ bản module trong Python là một file Python có phần mở rộng .py, chứa các đoạn mã Python như câu lệnh, các lớp, hàm, biến,... 

Ví dụ ***module random sẽ tương ứng với tệp random.py...***

Các chương trình sẽ được chia nhỏ thành các module nhỏ. 

Lợi ích của việc chia nhỏ này cung cấp cho người lập trình các cách thức để chia sẻ các hàm hoặc các đoạn mã có thể dùng chung một cách dễ dàng. 

Chúng ta chỉ cần đưa nó vào các module và sau đó gọi nó từ các file chương trình khác để sử dụng. 

Chúng ta cũng có thể tổng hợp nhiều modules thành một gói (hay package) giống như một thư mục gồm nhiều file khác nhau.

Đặc biệt là cách sử dụng nó thông qua việc ***import*** vào chương trình để có thể sử dụng các phương thức và tính năng trong nó.

### **Phân loại module**

***1. Các module được cung cấp sẵn (Built-in Modules).***



Đây là các module đi kèm với gói cài đặt Python mặc định từ trang chủ. Một trong những ưu điểm tuyệt với nhất của Python đó là nó cung cấp một thư viện hỗ trợ gồm rất nhiều các module, các hàm dựng sẵn phong phú làm cho việc lập trình trở lên nhẹ nhàng hơn rất nhiều.

 Một số module phổ biến có thể kể đến như:  ***datetime, os, math, sys, random***,...

***2. Các module do người dùng tự định nghĩa (User-defined Modules).***

Đây là những module khác được thêm vào do chính người dùng, hoặc cộng đồng đóng góp. Chúng ta có thể tạo các module và sử dụng trong chương trình hoặc chương trình khác. Các module này có thể chứa các lớp, các hàm, các biến,...theo đúng nhu cầu thiết kế của chúng ta.

Ngoài ra, chúng ta còn có các module do cộng đồng khắp thế giới cung cấp, được sử dụng rộng rãi như ***numpy, pandas, sklearn, tensorflow,...*** Thực chất đây cũng là các module do người dùng tự định nghĩa nhưng được chia sẻ và sử dụng rộng rãi trên thế giới.

### **Làm thế nào để import và sử dụng một module ?**

```python
import <module_name>
```

Khi trình thông dịch gặp một câu lệnh import nó sẽ thêm module với tên đằng sau import vào chương trình.

Module sẽ được tải vào duy nhất một lần, dù ta có cố gắng gọi nó nhiều lần ở các phần khác nhau của chương trình đi chăng nữa.

```python
import math

# Sử dụng hàm tính căn bậc hai trong module math
print(math.sqrt(5))
```

#### ***import cùng lúc nhiều module***

```python
# Import 2 modules
import math, random

print(math.factorial(5))
print(random.randint(10, 20))
```

Chúng ta có thể thêm từng module với từng lệnh import ở các dòng khác nhau, hoặc thêm tất cả các module muốn sử dụng ngay sau câu lệnh import, mỗi module sẽ phân tách nhau bởi dấu phẩy (,).

#### ***Sử dụng câu lệnh import và thay đổi lại tên của mô-đun***

```python
import math as m
print(m.sqrt(4))
```

Chúng ta đã đổi tên module math thành m. Điều này có thể giúp chúng ta tiết kiệm thời gian đánh máy trong một số trường hợp.

----
 Warning

Khi đổi tên như vậy thì tên math sẽ không được công nhận trong phạm vi của chương trình bên trên nữa do chúng ta đã gọi nó là m. Do đó, nếu trong thân chương trình chúng ta sử dụng ***math.sqrt*** sẽ ***không hợp lệ***, thay vào đó ta buộc phải sử dụng ***m.sqrt().***

----

#### ***Câu lệnh from…import để thêm một phần của module***

Đôi khi, chúng ta không muốn thêm toàn bộ các hàm hoặc các lớp trong một module, khi đó chúng ta có thể thêm các tên cụ thể từ một module mà không cần thêm toàn bộ module bằng cách sử dụng từ khóa ***from***.

```python
from math import sqrt
print(sqrt(4))
#       ^
#       |
```

----
![Alt text](https://tek4.vn/icons/tip-icon.svg) Tip&Trick

Lưu ý, với cách này, chúng ta sẽ không cần gọi hàm hoặc đối tượng của module theo tên module và dấu chấm như trước nữa. Chúng ta chỉ cần đơn giản gọi đúng tên hàm và đối tượng.

----

Ở đây, chúng ta chỉ thêm thành phần hàm sqrt từ mô-đun math. Ngoài ra, chúng ta cũng có thể thêm nhiều thành phần trong một mô-đun như sau.

```python
from math import sqrt, pi
print(sqrt(4))
print(pi)
```

Lưu ý rằng, nếu chúng ta thêm cái gì thì chúng ta chỉ sử dụng được cái đó, các hàm khác khi gọi mà không được thêm sẽ không hợp lệ!

```python
from math import sqrt
print(sqrt(4))
print(pi)
```

Tương tự như khi import toàn bộ module, chúng ta cũng có thể đổi tên của các đối tượng được import trong trường hợp import một phần của module.

```python
from math import sqrt as can
print(can(4))
```

#### ***Thêm tất cả các thành phần có trong mô-đun khi sử dụng from...import***

```python
from math import *
print(sqrt(9))
```

Ở trên, chúng ta thấy rằng việc import thiếu hàm hoặc đối tượng từ module sẽ làm chương trình báo lỗi, vì vậy để chắc ăn chúng ta có thể nhập tất cả các tên (các định nghĩa) từ một mô-đun bằng cách sử dụng dấu sao (*) như sau.

```python
from math import *
print(sqrt(9))
```

- Ở đây, chúng ta đã thêm tất cả các định nghĩa từ mô-đun math. 

- Tuy nhiên, việc thêm tất cả các thành phần bằng biểu tượng dấu sao (*) không phải là một cách lập trình tốt. 

- Điều này có thể dẫn đến các định nghĩa bị trùng lặp. Do các module có thể chứa những hàm có tên trùng nhau. 

- Vì vậy, thông thường nếu như chúng ta muốn thêm vào toàn bộ các thành phần của module, chúng ta nên ***sử dụng câu lệnh import*** và ***không dùng from*** như ở phần trên. Lí do, khi đó chúng ta sử dụng và chỉ định chính xác nó thuộc module nào thì chương trình sẽ không gây lỗi và nhập nhằng.
