## **Module trong Python**

Trong Python, cng nh a s cc ngn ng lp trnh khc, khi vit chng trnh, chng ta him khi ch s dng mt file m ngun duy nht. Trn thc t, chng trnh l tng hp ca rt nhiu file v th mc phc tp. Python qun l cc file v th mc ny di dng cc module v package.

V c bn module trong Python l mt file Python c phn m rng .py, cha cc on m Python nh cu lnh, cc lp, hm, bin,... 

V d ***module random s tng ng vi tp random.py...***

Cc chng trnh s c chia nh thnh cc module nh. 

Li ch ca vic chia nh ny cung cp cho ngi lp trnh cc cch thc  chia s cc hm hoc cc on m c th dng chung mt cch d dng. 

Chng ta ch cn a n vo cc module v sau  gi n t cc file chng trnh khc  s dng. 

Chng ta cng c th tng hp nhiu modules thnh mt gi (hay package) ging nh mt th mc gm nhiu file khc nhau.

c bit l cch s dng n thng qua vic ***import*** vo chng trnh  c th s dng cc phng thc v tnh nng trong n.

### **Phn loi module**

***1. Cc module c cung cp sn (Built-in Modules).***



y l cc module i km vi gi ci t Python mc nh t trang ch. Mt trong nhng u im tuyt vi nht ca Python  l n cung cp mt th vin h tr gm rt nhiu cc module, cc hm dng sn phong ph lm cho vic lp trnh tr ln nh nhng hn rt nhiu.

 Mt s module ph bin c th k n nh:  ***datetime, os, math, sys, random***,...

***2. Cc module do ngi dng t nh ngha (User-defined Modules).***

y l nhng module khc c thm vo do chnh ngi dng, hoc cng ng ng gp. Chng ta c th to cc module v s dng trong chng trnh hoc chng trnh khc. Cc module ny c th cha cc lp, cc hm, cc bin,...theo ng nhu cu thit k ca chng ta.

Ngoi ra, chng ta cn c cc module do cng ng khp th gii cung cp, c s dng rng ri nh ***numpy, pandas, sklearn, tensorflow,...*** Thc cht y cng l cc module do ngi dng t nh ngha nhng c chia s v s dng rng ri trn th gii.

### **Lm th no  import v s dng mt module ?**

```python
import <module_name>
```

Khi trnh thng dch gp mt cu lnh import n s thm module vi tn ng sau import vo chng trnh.

Module s c ti vo duy nht mt ln, d ta c c gng gi n nhiu ln  cc phn khc nhau ca chng trnh i chng na.

```python
import math

# S dng hm tnh cn bc hai trong module math
print(math.sqrt(5))
```

#### ***import cng lc nhiu module***

```python
# Import 2 modules
import math, random

print(math.factorial(5))
print(random.randint(10, 20))
```

Chng ta c th thm tng module vi tng lnh import  cc dng khc nhau, hoc thm tt c cc module mun s dng ngay sau cu lnh import, mi module s phn tch nhau bi du phy (,).

#### ***S dng cu lnh import v thay i li tn ca m-un***

```python
import math as m
print(m.sqrt(4))
```

Chng ta  i tn module math thnh m. iu ny c th gip chng ta tit kim thi gian nh my trong mt s trng hp.

----
 Warning

Khi i tn nh vy th tn math s khng c cng nhn trong phm vi ca chng trnh bn trn na do chng ta  gi n l m. Do , nu trong thn chng trnh chng ta s dng ***math.sqrt*** s ***khng hp l***, thay vo  ta buc phi s dng ***m.sqrt().***

----

#### ***Cu lnh fromimport  thm mt phn ca module***

i khi, chng ta khng mun thm ton b cc hm hoc cc lp trong mt module, khi  chng ta c th thm cc tn c th t mt module m khng cn thm ton b module bng cch s dng t kha ***from***.

```python
from math import sqrt
print(sqrt(4))
#       ^
#       |
```

Lu , vi cch ny, chng ta s khng cn gi hm hoc i tng ca module theo tn module v du chm nh trc na. Chng ta ch cn n gin gi ng tn hm v i tng.

----

 y, chng ta ch thm thnh phn hm sqrt t m-un math. Ngoi ra, chng ta cng c th thm nhiu thnh phn trong mt m-un nh sau.

```python
from math import sqrt, pi
print(sqrt(4))
print(pi)
```

Lu  rng, nu chng ta thm ci g th chng ta ch s dng c ci , cc hm khc khi gi m khng c thm s khng hp l!

```python
from math import sqrt
print(sqrt(4))
print(pi)
```

Tng t nh khi import ton b module, chng ta cng c th i tn ca cc i tng c import trong trng hp import mt phn ca module.

```python
from math import sqrt as can
print(can(4))
```

#### ***Thm tt c cc thnh phn c trong m-un khi s dng from...import***

```python
from math import *
print(sqrt(9))
```

 trn, chng ta thy rng vic import thiu hm hoc i tng t module s lm chng trnh bo li, v vy  chc n chng ta c th nhp tt c cc tn (cc nh ngha) t mt m-un bng cch s dng du sao (*) nh sau.

```python
from math import *
print(sqrt(9))
```

-  y, chng ta  thm tt c cc nh ngha t m-un math. 

- Tuy nhin, vic thm tt c cc thnh phn bng biu tng du sao (*) khng phi l mt cch lp trnh tt. 

- iu ny c th dn n cc nh ngha b trng lp. Do cc module c th cha nhng hm c tn trng nhau. 

- V vy, thng thng nu nh chng ta mun thm vo ton b cc thnh phn ca module, chng ta nn ***s dng cu lnh import*** v ***khng dng from*** nh  phn trn. L do, khi  chng ta s dng v ch nh chnh xc n thuc module no th chng trnh s khng gy li v nhp nhng.
