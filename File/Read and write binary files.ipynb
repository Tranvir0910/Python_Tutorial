## **Read and write binary files**

### **Làm việc với tệp nhị phân**

Đôi khi, bạn có thể cần phải làm việc với các tệp sử dụng chuỗi byte. 

Điều này được thực hiện bằng cách thêm ký tự **'b'** vào đối số **mode**. 

Tất cả các phương thức tương tự cho đối tượng tệp đều được áp dụng. Tuy nhiên, mỗi phương thức chấp nhận đầu vào khác nhau và trả về một đối tượng bytes:

Giả sử chúng ta có một tệp ảnh với tên là **a.png**. Chúng ta có thể mở tệp đó bằng Python và kiểm tra nội dung nội dung bên trong nó. 

Vì định dạng tệp **.png** được xác định rõ, tiêu đề của tệp là 8 byte được chia nhỏ như sau:

```python
with open('a.png', 'rb') as byte_reader:
    print(byte_reader.read(1))
    print(byte_reader.read(3))
    print(byte_reader.read(2))
    print(byte_reader.read(1))
    print(byte_reader.read(1))
```

### **Thực hiện ghi tệp nhị phân**

Trong ví dụ này, chúng ta sẽ mở một tệp bằng cách sử dụng **file = open (“document.bin”, ”wb”)** và sử dụng chế độ **“wb”** để ghi tệp nhị phân **document.bin**.

Chúng ta sẽ lấy một biến chứa một chuỗi ký tự và gán **“Python is easy to learn”**, Để thực hiện **chuyển đổi decode** câu đó, chúng ta sẽ sử dụng câu lệnh **bytearray(“Python is easy to learn” .encode (“ascii”))**. 

Và để viết chuỗi ký tự đó vào trong tệp, phương thức **file.write()** sẽ được sử dụng. Phương thức **write()** được sử dụng để ghi một đoạn đữ liệu được chỉ định vào trong một tệp. Và sau đó để đóng tệp, phương thức **close()** được gọi.

```python
file = open("document.bin","wb")
sentence = bytearray("Python is easy to learn".encode("ascii"))
file.write(sentence)
file.close()
```

Để đọc tệp ở dạng nhị phân, chúng ta cần sử dụng tệp **document.bin** đã được tạo có chứa nội dung trong đoạn mã bên trên và 
sử dụng chế độ **“rb”** để đọc tệp nhị phân. 

**document.bin** là tên của tệp. Phương thức **read()** được sử dụng để đọc tệp và trả về số byte được chỉ định từ tệp.

```python
file = open("document.bin","rb")
print(file.read(4))
file.close()
```

### **Thực hiện đọc một tệp nhị phân và lưu vào trong một mảng**

Trong ví dụ này, chúng ta sẽ mở một tệp với tên là **array.bin** và sử dụng chế độ **“wb”** để ghi tệp nhị phân. 

**array.bin** là tên của tệp. Sau đó, chúng ta thực hiện gán một mảng là **num = [2,4,6,8,10]**, và hàm **bytearray()** sẽ được sử dụng. 

Phương thức **bytearray()** trả về các đối tượng mảng byte. Để ghi mảng vào trong tệp, chúng ta cần sử dụng **file.write()** và **file.close()** để đóng tệp sau khi thao tác xong.

```python
file=open("array.bin","wb")
num=[2,4,6,8,10]
array=bytearray(num)
file.write(array)
file.close()
```

Để đọc mảng đã viết từ tệp bên trên, chúng ta sẽ sử dụng cùng một tệp, **file = open (“array.bin”, “rb”)**. 

Chế độ **“rb”** được sử dụng để đọc mảng từ tệp ở dạng nhị phân. 

Hàm **list()** được sử dụng để tạo đối tượng kiểu List hay danh sách **number = list (file.read (3))**. 

**file.read()** được sử dụng để đọc các byte từ tệp. 

**file.read(3)** được sử dụng để chỉ đọc ba giá trị số từ mảng. 

**file.close()** được sử dụng để đóng tệp sau khi đã đọc xong tệp.

```python
file=open("array.bin","rb")
number=list(file.read(3))
print (number)
file.close()
```
