## **Xử lý tệp trong Python**

Các File, hay còn được gọi là tệp, được sử dụng để lưu trữ các thông tin liên quan. 

Chúng được sử dụng để lưu trữ vĩnh viễn dữ liệu trong bộ nhớ (ví dụ: đĩa cứng). 

Vì bộ nhớ RAM rất dễ gây mất mát dữ liệu khi tắt máy tính, do vậy, chúng ta có thể sử dụng tệp để sử dụng dữ liệu về sau bằng cách lưu trữ chúng.

Khi chúng ta muốn đọc hoặc ghi vào một tệp, trước tiên chúng ta cần mở nó. Sau khi hoàn tất, nó cần phải được đóng lại để các tài nguyên gắn với tệp được giải phóng.

Như vậy, trong Python, các thao tác xử lý File trong Python diễn ra theo thứ tự sau:

- Mở tệp
- Đọc hoặc ghi tệp
- Đóng tệp

### **Mở tệp trong Python**

Python có một **hàm open()** đã được tích hợp sẵn để mở một tệp. Hàm này **trả về một đối tượng tệp**, còn được **gọi là handle**, đây là một đối tượng đặc biệt được sử dụng để đọc hoặc sửa đổi tệp.

```python
file = open("tek4vn.txt")
```

Theo mặc định là chế độ đọc ở dạng văn bản. Trong chế độ này, chúng ta sẽ nhận được chuỗi ký tự khi đọc từ tệp.

Mặt khác, chế độ nhị phân sẽ trả về dạng byte và đây là chế độ được sử dụng khi xử lý các tệp không phải văn bản như hình ảnh.

Khi làm việc với các tệp ở chế độ văn bản, ta sẽ chỉ định kiểu encoding cho dữ liệu.

**Ví dụ:**

```python
file = open("tek4vn.txt", mode = 'r', encoding = 'utf-8')
```

Ở câu lệnh này, chúng ta đang thực hiện việc mở tệp có tên là **"tek4vn.txt"** với chế độ **mode='r'** tức là mở tệp để đọc, và chỉ định kiểu encoding để đọc dữ liệu ra là **encoding='utf-8'**.

### **Đóng tệp trong Python**

#### **Sử dụng phương thức close()**

```python
file = open("tek4vn.txt", encoding = 'utf-8')
file.close()
```

Tuy nhiên, phương pháp này không hoàn toàn an toàn. 

Nếu một ngoại lệ xảy ra khi chúng ta đang thực hiện một số thao tác với tệp, đoạn mã sẽ kết thúc mà không đóng tệp. 

Tức là vẫn tồn tại phần tài nguyên đang xử lý tệp do chương trình đã gặp ngoại lệ và bị kết thúc cứng.

#### **Sử dụng khối lệnh xử lý ngoại lệ try...finally**

```python
try:
   file = open("tek4vn.txt", encoding = 'utf-8')
finally:
   file.close()
```

Bằng cách này, chúng ta có thể đảm bảo rằng tệp được đóng đúng cách ngay cả khi một ngoại lệ được đưa ra và khiến cho luồng thực thi của chương trình dừng lại. 

Chúng ta sẽ quay lại đoạn mã này ở chủ đề sau về xử lý lỗi vào ngoại lệ.
