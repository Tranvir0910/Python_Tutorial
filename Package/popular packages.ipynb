#### **Làm việc với thời gian qua module datetime**

Module datetime cung cấp các lớp để thao tác với thời gian ngày và giờ.

```python
from datetime import date
now = date.today()
print(now)
print(now.strftime("%m-%d-%y. %d %b %Y is a %A on the %d day of %B."))
```

#### **Đo lường hiệu năng thực thi với module timeit**

Module timeit được sử dụng để thực hiện đo lường hiệu suất thực thi của các thao tác trong chương trình.

```python
from timeit import Timer
Timer('t=a; a=b; b=t', 'a=1; b=2').timeit()
Timer('a,b = b,a', 'a=1; b=2').timeit()
```

#### **Làm việc với hệ điều hành thông qua module os**

```python
import os
os.getcwd()      #Trả về thư mục hiện tại
os.chdir('/server/accesslogs')   #Thay đổi thư mục
```

- getcwd() để lấy ra thư mục làm việc hiện tại.

- chdir(), thực hiện thay đổi thư mục làm việc hiện tại.
