## **Package trong Python**

Khi các chương trình thực tế đòi hỏi sự phức tạp với rất nhiều module, chúng ta sẽ đặt các module trong các gói (package) khác nhau.

Điều này làm cho một chương trình dễ quản lý và dễ hiểu hơn về mặt Logic. Một package là một thư mục chứa nhiều module và gói con.

Trong đó, thư mục này bắt buộc phải chứa một tệp có tên là **__ init__.py** để Python khi dịch chương trình có thể hiểu rằng nó là một gói Package. Tệp này có thể có thể để trống nhưng chúng ta thường đặt đoạn mã khởi tạo cho gói Package trong tệp này.

Chúng ta có thể thêm các mô-đun từ các gói bằng dấu chấm. Giả sử ta có một gói tên là Sinh_vien và các gói con của nó. Và trong các gói con, nó có chứa các module khác bên trong.

### **Làm việc với package trong Python**

Một package gồm nhiều module con bên trong, do đó, để làm việc với package trong Python, chúng ta cũng ***sử dụng câu lệnh import và from tương tự như khi làm việc với module.***

Ví dụ với gói Sinh_vien ở trên:

Để thêm module add từ gói trên vào chương trình, chúng ta sử dụng câu lệnh import như sau.

```python
import Sinh_vien.ID_sv.add
```

Tức là từ gói to cho đến gói con và cho đến module cụ thể giống như khi chúng ta truy cập vào một thư mục trên máy tính vậy.

Bây giờ, nếu mô-đun này chứa một hàm có tên có tên là ***add_ID()***, chúng ta phải sử dụng tên đầy đủ để tham chiếu đến nó.

```python
Sinh_vien.ID_sv.add.add_ID(100)
```

Cấu trúc này trông có vẻ dài dòng và khó nhìn, chúng ta có thể nhập module theo một cách khác là sử dụng thêm từ khóa from để chọn module muốn thêm như sau:

```python
from Sinh_vien.ID_sv import add
```

Function call

```python
add.add_ID(100)
```

Một cách khác để chỉ nhập hàm được yêu cầu (hoặc lớp hoặc biến) từ một mô-đun trong một gói Package như sau:

```python
from Sinh_vien.ID_sv.add import add_ID

add_ID(100)
```

Nguyên tắc là chúng ta import cái gì thì chúng ta phải dùng y như vậy và xuất phát từ nó để truy cập vào các thành phần con bên trong.

### **Đường dẫn tìm kiếm các module và package**

Khi chúng ta tiến hành thêm một module, Python sẽ xem xét một số vị trí để tìm kiếm module này. 

Trình thông dịch trước tiên sẽ tìm kiếm một module được tích hợp sẵn trong thư mục cài đặt. 

Sau đó (nếu không tìm thấy module được tích hợp sẵn nào có tên được yêu cầu), Python sẽ xem xét danh sách các thư mục được xác định trong **sys.path** được khởi tạo từ biến môi trường **PYTHONPATH** của hệ thống.

```python
import sys
print(sys.path)
```

Nếu không tìm thấy trong tất cả các đường dẫn này, thì chương trình sẽ báo lỗi do không tìm thấy module được yêu cầu.

#### **Tóm lại:**

- Package là các gói chứa nhiều module khác nhau.

- Để sử dụng gói chúng ta cũng sử dụng các câu lệnh import, from và as tương tự như với module

- Quy tắc sử dụng đó là chúng ta import như đọc các thư mục và file, các đường dẫn sẽ được phân tách bởi dấu . cho đến vị trí muốn đọc để sử dụng
Các module hay package sẽ được ưu tiên sử dụng ở thư mục làm việc của chương trình trước, sau đó mới được tìm kiếm trong các thư mục được chỉ định trong biến môi trường.
